#**8. Exploratory Data Analysis - Time Series Analysis (Absolute Values of KPIs)**

We’ll analyse: popularity, energy, danceability, valence — yearly averages, trends, YoY growth, and smooth trends (rolling mean).

```python
# 1. KPI selection and yearly aggregation
kpis = ['popularity', 'energy', 'danceability', 'valence']
yearly = df.groupby('year')[kpis].mean().reset_index().sort_values('year')
yearly.head()
```

Explanation: creates yearly DataFrame with mean KPI per year.

```python
# 2. Plot yearly KPI trends (same y-scale)
plt.figure(figsize=(14,6))
for col in kpis:
    sns.lineplot(data=yearly, x='year', y=col, marker='o', label=col)
plt.title('Yearly Average of KPIs (popularity, energy, danceability, valence)')
plt.xlabel('Year')
plt.ylabel('Mean value')
plt.xticks(rotation=45)
plt.legend(title='KPI')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# 3. Z-score scaling so shapes compare (if magnitudes differ)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_values = scaler.fit_transform(yearly[kpis].values)
yearly_scaled = yearly[['year']].copy()
yearly_scaled[kpis] = scaled_values

plt.figure(figsize=(14,6))
for col in kpis:
    sns.lineplot(data=yearly_scaled, x='year', y=col, marker='o', label=col)
plt.title('Yearly KPI Trends (z-score scaled)')
plt.xlabel('Year')
plt.ylabel('Z-score')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 4. Rolling averages for smoothing
rolling_df = yearly.set_index('year').copy()
rolling_df['pop_3yr'] = rolling_df['popularity'].rolling(3, center=True, min_periods=1).mean()
rolling_df['pop_5yr'] = rolling_df['popularity'].rolling(5, center=True, min_periods=1).mean()

plt.figure(figsize=(12,5))
sns.lineplot(data=rolling_df.reset_index(), x='year', y='popularity', label='yearly mean', marker='o')
sns.lineplot(data=rolling_df.reset_index(), x='year', y='pop_3yr', label='3-yr rolling', linestyle='--')
sns.lineplot(data=rolling_df.reset_index(), x='year', y='pop_5yr', label='5-yr rolling', linestyle=':')
plt.title('Popularity — yearly mean and rolling averages')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 5. YoY absolute and percentage change for each KPI
yoy = yearly.copy()
for col in kpis:
    yoy[f'{col}_yoy_abs'] = yoy[col].diff()
    yoy[f'{col}_yoy_pct'] = yoy[col].pct_change() * 100
# Show most recent 8 years for context
yoy[['year'] + [c for c in yoy.columns if 'yoy' in c]].tail(8).round(3)
```

```python
# 6. Biggest YoY increases and decreases (percentage) for each KPI
report = {}
for col in kpis:
    col_pct = f'{col}_yoy_pct'
    best_up = yoy.loc[yoy[col_pct].idxmax(), ['year', col_pct]]
    best_down = yoy.loc[yoy[col_pct].idxmin(), ['year', col_pct]]
    report[col] = {'biggest_increase': best_up.to_dict(), 'biggest_decrease': best_down.to_dict()}

import pprint
pprint.pprint(report)
```

# **9) Exploratory Data Analysis - Time Series Analysis (Growth of KPIs)**

Examine year-over-year growth trends of major KPIs to identify shifts and performance changes over time.

```python
# 1️⃣ Compute yearly averages for KPIs
kpis = ['popularity', 'energy', 'danceability', 'valence']
yearly_growth = df.groupby('year')[kpis].mean().reset_index().sort_values('year')
yearly_growth
```

```python
# 2️⃣ Calculate YoY growth (% change from previous year)
for col in kpis:
    yearly_growth[f'{col}_growth_pct'] = yearly_growth[col].pct_change() * 100

yearly_growth.tail(10)  # view recent years
```

```python
# 3️⃣ Plot Year-over-Year Growth (%)
plt.figure(figsize=(14,6))
for col in kpis:
    sns.lineplot(data=yearly_growth, x='year', y=f'{col}_growth_pct', marker='o', label=col)
plt.axhline(0, color='gray', linestyle='--')
plt.title('Year-over-Year Growth (%) of Key KPIs')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend(title='KPI')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# 4️⃣ Find years with biggest positive and negative growth
growth_report = {}
for col in kpis:
    growth_report[col] = {
        'max_growth': yearly_growth.loc[yearly_growth[f'{col}_growth_pct'].idxmax(), ['year', f'{col}_growth_pct']].to_dict(),
        'max_decline': yearly_growth.loc[yearly_growth[f'{col}_growth_pct'].idxmin(), ['year', f'{col}_growth_pct']].to_dict()
    }
import pprint
pprint.pprint(growth_report)
```

🎯 Insights — Time Series Growth of KPIs

• Popularity growth shows major spikes around key industry periods (e.g., 2000s digital rise, 2020s streaming boom).  
• Energy and Danceability often move together — fast-growing periods reflect shifts toward upbeat, high-BPM genres.  
• Valence (positivity) growth fluctuates — dips may correspond to trends toward moodier or darker tones.  
• Sharp negative growth years can signal data bias (few tracks or incomplete records) — handle carefully.  
• Sustained positive growth across multiple KPIs indicates evolving listener preference toward energetic, danceable, and emotionally bright music.

# **12) Outlier Analysis**
Detect and interpret outliers using statistical methods like the IQR approach to understand extreme or unusual cases.

```python
# 1️⃣ Choose important numerical features
num_cols = ['popularity', 'duration_ms', 'energy', 'danceability', 'valence', 'loudness']

df_outlier = df[num_cols].copy()
df_outlier.describe()
```

```python
# 2️⃣ Define function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

```python
# 3️⃣ Boxplots to visually detect outliers
# Improved boxplots: duration in minutes (log scale) + others on linear scale
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

cols = ['popularity','duration_ms','energy','danceability','valence']
plot_df = df[cols].copy()

# Convert duration to minutes
plot_df['duration_min'] = plot_df['duration_ms'] / 60000.0

# Prepare subplots: duration on left (log scale), others on right
fig, axes = plt.subplots(1, 5, figsize=(18,5), gridspec_kw={'width_ratios':[1.2,1,1,1,1]})

# 1) Duration (log scale) to tame extreme long tracks
sns.boxplot(y=plot_df['duration_min'].dropna(), ax=axes[0], color='lightcoral',
            flierprops={'marker':'o','markeredgecolor':'gray','markersize':4, 'alpha':0.6})
axes[0].set_ylabel('Duration (minutes)')
axes[0].set_yscale('log')
axes[0].set_title('Duration (log minutes)')

# 2-5) Other KPIs (linear)
other = ['popularity','energy','danceability','valence']
for ax, col in zip(axes[1:], other):
    sns.boxplot(y=plot_df[col].dropna(), ax=ax, color='lightblue',
                flierprops={'marker':'o','markeredgecolor':'gray','markersize':4, 'alpha':0.6})
    ax.set_title(col)
    ax.set_ylabel('')  # keep only duration with y-label for clarity

plt.suptitle('Boxplots of Key Features (duration on log-minutes scale)')
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
# 4️⃣ Detect and count outliers in each numerical feature
for col in df_outlier.columns:
    outliers, lb, ub = detect_outliers_iqr(df_outlier, col)
    print(f"{col}: {len(outliers)} outliers | Lower Bound = {lb:.2f}, Upper Bound = {ub:.2f}")
```

```python
# 5️⃣ Example: Outliers in track duration
outliers_duration, lb, ub = detect_outliers_iqr(df_outlier, 'duration_ms')
print(f"Outliers in duration_ms: {len(outliers_duration)} tracks")

# Convert to minutes for readability
outliers_duration['duration_min'] = outliers_duration['duration_ms'] / 60000
display(outliers_duration[['popularity', 'duration_min', 'energy', 'danceability', 'valence']].head(10))
```

🎯 **Insights — Outlier Analysis**

• Several extreme values were detected across key features such as Duration, Loudness, and Popularity.  
• Duration outliers (very long tracks) likely represent podcasts, live recordings, or extended mixes, not standard songs.  
• Loudness outliers on the quieter end suggest poorly normalized or acoustic tracks; extremely loud ones indicate aggressive mastering.  
• Popularity outliers (scores >80) are rare — only a small fraction of tracks achieve high audience engagement.  
• Energy and Danceability have few outliers, showing consistent production patterns across most songs.  
• Outliers should be handled contextually — remove only when they distort analysis; otherwise, they may highlight special cases worth studying.

# **14) Future Scope / Limitations**
Discuss the limitations of the dataset or analysis and suggest possible extensions for future exploration.



## ⚠️ Limitations of the Dataset and Analysis

1. **Incomplete Metadata:**  
   The dataset does not include key contextual variables like **genre**, **artist popularity growth**, or **playlist inclusion** — all of which could have a major influence on track success.

2. **Popularity Bias:**  
   The popularity score on Spotify is **dynamic and region-dependent**. It changes with user activity, meaning the current values may not reflect long-term trends.

3. **Missing Listener Demographics:**  
   Without information on **age, region, or listener behavior**, we can’t link musical attributes to audience preferences directly.

4. **Static Snapshot of Data:**  
   The dataset likely represents a **single time frame**, not continuous updates. Therefore, **temporal evolution** of music trends (e.g., how energy levels changed across years) remains partially unexplored.

5. **Data Imbalance:**  
   Some languages or styles have very few entries (like Malayalam, Telugu), which can distort statistical conclusions and visualizations.

6. **Correlation, Not Causation:**  
   Observing that *high energy tracks tend to be more popular* does **not imply causation** — external factors like promotion, social media trends, or artist fame also play key roles.
